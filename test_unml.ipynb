```python
import numpy as np
import pandas as pd
import unml
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"cul_cat1.csv",index_col=[0])
col=['peak_freq','log_bc_width','log_flux','log_fluence','redshift','fre_width','log_in_duration','log_energy', 'log_luminosity','log_T_B']
data = df[col]
```

```python
dr=unml.dimensionality_reduction(data=data,standardscale='True',method='PCA',n_components=2)
plt.scatter(dr[:,0],dr[:,1])
```
